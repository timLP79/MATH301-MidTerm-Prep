```python
from sympy import linsolve, symbols, Eq, Matrix, S
```

Equation example

```python
x1, x2, x3 = symbols('x1, x2, x3')
eq1 = Eq(x1 + 2 * x2 - x3, 6)
eq2 = Eq(-3 * x1 + 3 * x3, 10)
linsolve((eq1, eq2), (x1, x2, x3))
```

Augmented Matrix Example

```python
x, y, z = symbols('x, y, z')
A = Matrix(([1, 2, -1, 6], [-3, 0, 3, 10]))
linsolve(A, (x, y, z))
```
